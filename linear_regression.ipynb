```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(180)

X = np.linspace(-10,10,500)
y = 2*X + 7 + np.random.normal(0,2,size=X.shape)
plt.scatter(X,y)
plt.show()
```

```python
split_ratio = 0.8
split_index = int(split_ratio * len(X))
X_train , X_test = X[:split_index], X[split_index:]
y_train , y_test = y[:split_index], y[split_index:]
```

```python
class LinearRegression:
  def __init__(self,lr = 0.01,epochs = 100):
    self.lr = lr
    self.epochs = epochs
    self.slope = 0
    self.intercept = 0

  def predict(self,X):
    return self.slope * X + self.intercept

  def compute_mse(self,y_true,y_pred):
    return np.mean((y_true - y_pred)**2)

  def compute_rmse(self,y_true,y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

  def fit(self,X,y):
    m = len(X)
    for _ in range(self.epochs):
      # Prediction
      y_pred = self.predict(X)

      # Error Computation
      error = (y_pred - y)

      # Gradient Computation
      d_slope = (2/m) * np.sum(X*error)
      d_intercept = (2/m) * np.sum(error)

      # weight updation
      self.slope -=self.lr * d_slope
      self.intercept -= self.lr * d_intercept

      print("Epoch -> ",_," MSE Error- > ",self.compute_mse(y,y_pred))


```

```python
model = LinearRegression(0.01,200)
model.fit(X_train,y_train)
```
