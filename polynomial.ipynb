```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('data_tem.csv')
```

```python
df
```

```python
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.scatter(df[['Temperature']],df['Pressure'],marker='*')
```

```python
from sklearn import linear_model
```

```python
rg = linear_model.LinearRegression()
rg.fit(df[['Temperature']],df['Pressure'])
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=6)
x_poly = poly.fit_transform(df[['Temperature']])
poly.fit(x_poly,df['Pressure'])
rg2 = linear_model.LinearRegression()
rg2.fit(x_poly,df['Pressure'])
```

```python
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.scatter(df[['Temperature']],df['Pressure'],marker= "+")
plt.plot(df['Temperature'],rg.predict(df[['Temperature']]),color='red')
plt.plot(df['Temperature'],rg2.predict(poly.fit_transform(df[['Temperature']])),color='green')
```
